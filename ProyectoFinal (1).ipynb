```python
df.plot(kind='line')
plt.yticks(fontsize = 17)
plt.title('Grafica de Variables', fontsize = 17)
plt.show()
```

```python
sns.histplot(df.price)
plt.title('Histograma Variable: Precio', fontsize = 17)
```

```python
sns.distplot(df.price)
plt.title('Histograma Variable con regresion: Precio', fontsize = 17)
```

```python
sns.histplot(df.displacement)
plt.title('Histograma Variable: Distancia de Desplazamiento', fontsize = 17)
```

```python
sns.distplot(df.displacement)
plt.title('Histograma Variable con regresion: Distancia de Desplazamiento', fontsize = 17)
```

```python
'''Comparacion de Histogramas precio vs desplazamiento'''
```

```python
from matplotlib.pyplot import figure
figure(figsize=(10,5),dpi=70)
plt.subplot(1,2,1)

plt.hist('price',data = df,color = 'b')
plt.title('price')
plt.subplot(1,2,2)
plt.hist('displacement',data = df,color = 'r')
plt.title('displacement')
          
plt.show()
```

```python
'''Diagram Scatter Plot: Precio & Displacement'''
```

```python
df.plot(kind = 'scatter',
y = 'price',
x = 'displacement')
plt.title('Diagrama de precio vs Desplazamiento')
plt.show()
```
